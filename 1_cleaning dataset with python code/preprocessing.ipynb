```python
dset3 = pd.read_excel("crop yield prediction using machine learning.xlsx")
dset3.head(10)
```

```python
dset3.info()
```

```python
dset3.drop(['State_Name', 'District_Name', 'Crop_Year'], axis = 1, inplace = True)
dset3.head()
```

```python
missing_in_dset3 = dset3.isnull()
missing_in_dset3
```

```python
headers = ['Season', 'Crop', 'Temperature', 'humidity', 'soil moisture', ' area', 'Production']
for column in headers:
    print(missing_in_dset3[column].value_counts())
    print("")
```

```python
# simply drop whole row with NaN in "price" column
dset3.dropna(subset=['Production'], inplace=True, axis=0)

# reset index, because we droped two rows
dset3.reset_index(drop=True, inplace=True)
```

```python
missing_in_dset3 = dset3.isnull()
headers = ['Season', 'Crop', 'Temperature', 'humidity', 'soil moisture', ' area', 'Production']
for column in headers:
    print(missing_in_dset3[column].value_counts())
    print("")
```
